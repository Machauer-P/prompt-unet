# PromptUNet Test Pipeline Runner
This notebook allows you to configure and run the modularized test pipeline for PromptUNet models natively. <br>
Just copy, rename and test new/other models.

```python
import os
import sys

sys.path.append(os.path.join(os.getcwd(), '..', '..'))

from utils.augmentations import PromptUNetAugmenter
from evaluation.eval_prompt_unet.eval_p_unet_pipeline import PromptUNetTester
```

### 1. Configure Testing Parameters

```python
dimensions = ['x', 'y', 'z'] 
offsets = [5, 10, 20]
models = [
    'p_unet_21.keras',
    'p_unet_272.keras',
    'p_unet_273.keras',
    'p_unet_282.keras',
    'p_unet_283.keras'
]


max_data_points = 50
models_dir = "training/"

# You can define active augmentations here. Example: augmentations = [{'name': 'add_noise'}]
augmentations = []
```

### 2. Run Pipeline

#### MRI

```python
dataset_path = ["data/test_data/HanSeg_MRI.pkl"]

# Initialize tester
tester = PromptUNetTester(
    dataset_path=dataset_path,
    models_dir=models_dir,
    augmentations=augmentations,
    max_data_points=max_data_points
)

# Execute full validation loop (outputs and plots will be rendered sequentially)
results = tester.run_pipeline(
    dimensions=dimensions, 
    offsets=offsets, 
    models=models, 
    threshold=0.45, 
    num_visualize=2
)
```

#### CT

```python
dataset_path = ["data/test_data/HanSeg_CT.pkl"]

# Initialize tester
tester = PromptUNetTester(
    dataset_path=dataset_path,
    models_dir=models_dir,
    augmentations=augmentations,
    max_data_points=max_data_points
)

# Execute full validation loop (outputs and plots will be rendered sequentially)
results = tester.run_pipeline(
    dimensions=dimensions, 
    offsets=offsets, 
    models=models, 
    threshold=0.45, 
    num_visualize=2
)
```
